```python
import pandas as pd
from textblob import TextBlob
```

```python
# Load your Excel file from Google Drive
file_path = '/content/drive/My Drive/Data Analyst - Test Data - US (1).csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')
```

```python
df.head()
```

```python
# Modify the function to handle non-string values
def get_sentiment(text):
    if isinstance(text, str):
        analysis = TextBlob(text)
        # Return sentiment polarity (negative, neutral, positive)
        if analysis.sentiment.polarity > 0:
            return 'Positive'
        elif analysis.sentiment.polarity == 0:
            return 'Neutral'
        else:
            return 'Negative'
    else:
        return 'No Review'
```

```python

# Apply sentiment analysis
df['Sentiment'] = df['Review'].apply(get_sentiment)
```
